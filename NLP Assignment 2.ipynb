```python
from nltk.tokenize import sent_tokenize,word_tokenize
import re
from nltk.util import ngrams
from operator import itemgetter
import math
import numpy as np


################### Sentence Tokenization ###################

f = open(r"C:\Users\Dhananjay Sonawane\Desktop\alice1.txt","r")              # Provide path of book here
contents = f.read()
temp = sent_tokenize(contents)
allSent = []
listTokens = []
reg = '[A-Za-z0-9]*'
wordFreq = {'__start__':0,'__end__':0}
temp2 = []
for sentence in temp:
    sentence = sentence.lower()
    wordList = word_tokenize(sentence)                                        # Filtering unneccessary data
    set1 = '__start__ '
    t = wordFreq['__start__']
    wordFreq['__start__'] = t + 1
    
    for word in wordList:
        token = re.match(reg,word)
        str1 = token.group(0)
        if str1 != '':
            #print(token)
            listTokens.append(str1.lower())
            set1 += str1.lower()
            set1 += ' '
    set1 = set1.strip() + ' __end__'
    t = wordFreq['__end__']
    wordFreq['__end__'] = t + 1
    allSent.append(set1)                                                      # Filtered sentence list
#print(allSent)
    
################### Dividing data set ###################

lenAllSen = len(allSent)
upto = int(0.8 * lenAllSen)
trainingData = allSent[:upto]
testData = allSent[upto:]
    
'''print(len(trainingData))
print('-----------------------------------------------------')
print(len(testData))'''

print('################### MLE for Unigram ###################')
print()

tokens = len(listTokens)

for word in listTokens:
    if word not in wordFreq.keys():
        wordFreq[word] = listTokens.count(word)
#print(wordFreq)
for k,v in reversed(sorted(wordFreq.items(), key = itemgetter(1))):
    print('p( ' +k+' ) = ' + str(v/tokens) )
    
    
```

```python
print('################### MLE for Bigram ###################')
print()
bigramDict = {}
for sentence in trainingData:
    #print(sentence)
    #sentence = sentence.lower()
    #len_sen = len(sentence)
    wrdList = ngrams(word_tokenize(sentence),2)
    
    for bigram in wrdList:
        if bigram not in bigramDict:
            bigramDict[bigram] = 1
        else:
            v = bigramDict[bigram]
            bigramDict[bigram] = v + 1
#print(bigramDict)
for k,v in reversed(sorted(bigramDict.items(), key = itemgetter(1))):
    print('p( ' +k[1] + ' / '+ k[0] +' ) = ' + str(v/wordFreq[k[0]]) )
```

```python
print('################### MLE for Trigram ###################')
print()
trigramDict = {}
for sentence in trainingData:
    sentence = sentence.lower()
    len_sen = len(sentence)
    wrdList = ngrams(word_tokenize(sentence),3)
    
    for trigram in wrdList:
        if trigram not in trigramDict:
            trigramDict[trigram] = 1
        else:
            v = trigramDict[trigram]
            trigramDict[trigram] = v + 1
#print(bigramDict)
for k,v in reversed(sorted(trigramDict.items(), key = itemgetter(1))):
    tup = (k[0].strip(),k[1].strip())
    print('p( ' +k[2] + ' / '+ k[0],k[1] +' ) = ' + str(v/bigramDict[tup]) )
```

```python
print('############ MLE for Quadgram ############')
print()

quadDict = {}
for sentence in trainingData:
    sentence = sentence.lower()
    len_sen = len(sentence)
    wrdList = ngrams(word_tokenize(sentence),4)
    
    for quadgram in wrdList:
        if quadgram not in quadDict:
            quadDict[quadgram] = 1
        else:
            v = quadDict[quadgram]
            quadDict[quadgram] = v + 1
#print(quadgramDict)
for k,v in reversed(sorted(quadDict.items(), key = itemgetter(1))):
    tup = (k[0].strip(),k[1].strip(),k[2].strip())
    print('p( ' +k[3] + ' / '+ k[0],k[1],k[2] +' ) = ' + str(v/trigramDict[tup]) )
```

```python

print('################ Calculate probalility of sentence ################')
print()
print('Probability of sentence is calculated in logarithmic scale')
print()
#print(trainingData)
#print(testData)
def probUnigram(wr):
    return wordFreq[wr]/tokens

def probBigram(tup1):
    return bigramDict[tup1]/wordFreq[tup1[0]]

def probTrigram(tup1):
    return trigramDict[tup1]/bigramDict[tup1[:2]]

def probQuadgram(tup1):
    return quadDict[tup1]/trigramDict[tup1[:3]]

def probability(sentence,model):
    
    err = 'One of the word in given sentence is not trained in '+ model +' model of corpus. Hence,'
    
    wrds = sentence.split()

    logProb = 0
    normalProb = 1
    if model=='unigram':
        for wr in wrds:
            try:
                logProb += math.log(probUnigram(wr))
            except KeyError:
                print(err)
                return 0
            
    elif model == 'bigram':
        last = 2
        for i in range(2,len(wrds)):
            l2 = wrds[i-last : i]
            tup1 = tuple(l2)
            #print('prob of '+str(tup1)+' is '+str(probBigram(tup1)))
            try:
                logProb += math.log(probBigram(tup1))
                #normalProb *= probBigram(tup1)
            except KeyError as e:
                print(err)
                return 0
            
    elif model == 'trigram':
        last = 3
        for i in range(3,len(wrds)):
            lTri = wrds[i-last : i]
            tupTri = tuple(lTri)
            try:
                logProb += math.log(probTrigram(tupTri))
            except KeyError as e:
                print(err)
                return 0
            
    elif model == 'quadgram':
        last = 4
        for i in range(4,len(wrds)):
            lQuad = wrds[i-last : i]
            tupQuad = tuple(lQuad)
            try:
                logProb += math.log(probQuadgram(tupQuad))
            except KeyError as e:
                print(err)
                return 0
    
    return logProb

sen1 = trainingData[200]
model = 'unigram'
calc = probability(sen1,model)
print('p('+ sen1 +' ) in '+ model +' model :'  + str(calc))
print()

model = 'bigram'
calc = probability(sen1,model)
print('p('+ sen1 +' ) in '+ model +' model :'  + str(calc))
print()

model = 'trigram'
calc = probability(sen1,model)
print('p('+ sen1 +' ) in '+ model +' model :'  + str(calc))
print()

model = 'quadgram'
calc = probability(sen1,model)
print('p('+ sen1 +' ) in '+ model +' model :'  + str(calc))
print()

print('**** Notice that for same sentence probalibity increases as "n" in n-gram model increases')
print()

sen1 = trainingData[110]
model = 'bigram'
calc = probability(sen1,model)
print('p('+ sen1 +' ) in '+ model +' model :'  + str(calc))
print()

sen1 = trainingData[350]
model = 'bigram'
calc = probability(sen1,model)
print('p('+ sen1 +' ) in '+ model +' model :'  + str(calc))
print()
    
    
        
```

```python
print('###################### Add-1 Smoothing #################')
print()
print('Before smoothing :')
before_smoothing = {}
for i in wordFreq.keys():
    for j in wordFreq.keys():
        if (i,j) in bigramDict.keys():
            before_smoothing[(i,j)] = bigramDict[(i,j)] / wordFreq[i]
        else:
            before_smoothing[(i,j)] = 0
        #print(i,j + ' : '+ str(before_smoothing[(i,j)]))
        '''Uncomment this print to see probability of 
           every possible bigram

           This might give memory error due to large
           number of possible bigram'''
```

```python
print('After smoothing :')         
after_smoothing = {}
for i in wordFreq.keys():
    for j in wordFreq.keys():
        
        if (i,j) in bigramDict.keys():
            after_smoothing[(i,j)] = (bigramDict[(i,j)] + 1) / (wordFreq[i] + vocabLen) 
        else:
            after_smoothing[(i,j)] = 1 / (wordFreq[i] + vocabLen)
        #print(i,j + ' : '+ str(after_smoothing[(i,j)]))                     
        '''Uncomment this print to see probability of 
          every possible bigram

          This might give memory error due to large
          number of possible bigram'''
```

Drastic change in probability occurs for particular bigram (w_i-1 , w_i) whose occurance in corpus is high but occurance 
of most of bigram (w_i-1 , x) , x belongs to vocabulary is 0. Some of probability mass function of (w_i-1 , w_i) will get 
distributed over (w_i-1 , x) , x belongs to vocabulary.

```python
print('################# Good Turing ##################')
print()
freqOfFreq = {}
for k,v in wordFreq.items():
    if v not in freqOfFreq.keys():
        freqOfFreq[v] = 1
    else:
        freqOfFreq[v] += 1

ky = freqOfFreq.keys()
key_c = {}
print('Count c \t Good Turing effective count (c*) \t\t Difference')
for i in range(max(freqOfFreq.keys())):
    if (i in ky) and (i+1 in ky):
        c = (i+1) * freqOfFreq[i+1] / freqOfFreq[i]
        if i not in key_c.keys():
            key_c[i] = c
        if i< 10:
            print(i,'\t\t',round(c,5),'\t\t\t\t\t',round(abs(i-c),5))

print()
print('Constant distance value "d" is: ',0.9)
```

```python
print('################# Perplexity Using Add-1 smoothing #################')
print()
tstBigramDict ={}

for sentence in testData:
    tstWrdLst = ngrams(word_tokenize(sentence),2)
    #print(dir(tstWrdLst))
    mul = 1
    for bigram in tstWrdLst:
        mul *= after_smoothing[bigram]
        if bigram not in tstBigramDict:
            tstBigramDict[bigram] = 1
        else:
            v = tstBigramDict[bigram]
            tstBigramDict[bigram] = v + 1
    try:
        mul = mul**(-1/(len(sentence.split())))
        print('perplexity( ' + sentence+ ' ) : ' + str(mul))
    except:
        pass

```

```python
print('################# Perplexity Using Good Turing smoothing #################')
print()
reg = '[A-Za-z0-9]*'
tstListToken = []

for sen3 in testData:
    allWords = word_tokenize(sen3)
    
    for word in allWords:
        token = re.match(reg,word)
        str1 = token.group(0)
        if str1 != '':
            #print(token)
            tstListToken.append(str1.lower())
noTokens = len(tstListToken)

per_good = {}
for k,v in wordFreq.items():
    if k not in per_good:
        try:
            per_good[k] = key_c[v] / noTokens
        except:
            pass
#print(per_good)
for sentence in testData:
    tstWrdLst = ngrams(word_tokenize(sentence),2)
    #print(dir(tstWrdLst))
    mul = 1
    for bigram in tstWrdLst:
        try:
            mul *= per_good[bigram[1]]
        except:
            pass
        
    try:
        mul = mul**(-1/(len(sentence.split())))
        print('perplexity( ' + sentence+ ' ) : ' + str(mul))
    except:
        pass
```
